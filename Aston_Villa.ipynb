# **Aston Villa 23/24**

## **Load Data**


```python
import pandas as pd

# URL should be a string, enclosed in quotes
url = "https://raw.githubusercontent.com/jry50390/Aston-Villa-PL-23-24/main/FILE%20PREPARED"
df = pd.read_csv(url)
df
```

## **Data Preparation**

### Data separatation as X and Y

```python
y = df['Points']
y
```

```python
x = df.drop('Points', axis=1)
x
```

### Data splitting

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=20)
```

```python
x_test
```

## **Model Building**

### **Linear Regression**

**Training Model**


```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(x_train, y_train)
```

**Applying the model to make a prediction**

```python
y_lr_train_pred = lr.predict(x_train)
y_lr_test_pred = lr.predict(x_test)
```

```python
y_lr_train_pred
```

```python
y_lr_test_pred
```

**Evaluate Model Performance**

```python
from sklearn.metrics import mean_squared_error, r2_score
lr_train_mse = mean_squared_error(y_train, y_lr_train_pred)
lr_train_r2 = r2_score(y_train, y_lr_train_pred)

lr_test_mse = mean_squared_error(y_test, y_lr_test_pred)
lr_test_r2 = r2_score(y_test, y_lr_test_pred)
```

```python
lr_train_mse
```

```python
lr_train_r2
```

```python
lr_test_mse
```

```python
lr_test_r2
```

```python
lr_results = pd.DataFrame(['Linear Regression:', lr_train_mse, lr_train_r2, lr_test_mse, lr_test_r2]).transpose()
lr_results.columns = ['Method', 'MSE Training','R2 Training', 'MSE Test', 'R2 Test']
```

```python
lr_results
```

### **Random Forest**

### **Training the model**

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(max_depth=2, random_state=20)
rf.fit(x_train, y_train)
```

**Applying the model to make a prediction**

```python
y_rf_train_pred = rf.predict(x_train)
y_rf_test_pred = rf.predict(x_test)
```

**Evaluate Model Performance**

```python
from sklearn.metrics import mean_squared_error, r2_score
rf_train_mse = mean_squared_error(y_train, y_rf_train_pred)
rf_train_r2 = r2_score(y_train, y_rf_train_pred)

rf_test_mse = mean_squared_error(y_test, y_rf_test_pred)
rf_test_r2 = r2_score(y_test, y_rf_test_pred)
```

```python
rf_results = pd.DataFrame(['Random Forest:',rf_train_mse, rf_train_r2, rf_test_mse, rf_test_r2]).transpose()
rf_results.columns = ['Method', 'MSE Training','R2 Training', 'MSE Test', 'R2 Test']
rf_results
```

### **Model Comparison**

```python
df_models = pd.concat([lr_results, rf_results], axis=0)
df_models
```

```python
df_models.reset_index(drop=True)
```
